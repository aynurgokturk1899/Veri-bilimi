plt.plot fonksiyonu

```python
# modellediğimiz ilişki fonksiyon f(x)=2x
x =[0,1,2,3,4,5,6]
y =[0,2,4,6,8,10,12]

plt.figure(figsize=(10,8))  # ne kadar küçültürseniz o kadar daha detay görebilirsiniz
plt.plot(x,y)
plt.show()
```

```python
# -3x grafiğini gösteren çizgi grafiği çiziniz
x=[0,1,2,3,4,5]
y=[0,-3,-6,-9,-12,-15]
plt.figure(figsize=(4,6))
plt.plot(x,y,'--', color='red')
```

```python
x=np.arange(6)
y=4*x-2
plt.figure(figsize=(4,6))
plt.plot(x,y,'--',color='blue')
```

```python
# Lineer olmayan fonksiyonlar
```

```python
x=np.arange(100)/10
y=np.sin(x)
plt.plot(x,y)
```

```python
x=np.arange(6)
y1=x*2
y2=(x**2)/10

plt.plot(x,y1,label="2*x")
plt.plot(x,y2,label="x**2")

plt.legend(loc='upper right')
plt.show()
```

plt.scatter() fonksiyonu

```python
x =[0,1,2,3,4,5,6]
y =[0,2,4,6,8,10,12]

plt.figure(figsize=(10,8))  # ne kadar küçültürseniz o kadar daha detay görebilirsiniz
plt.plot(x,y,color='red')
plt.scatter(x,y)
plt.show()
```

```python
# numpy kullanarak cos(x) ve sin(x) fonksiyonunu
```

```python
x=np.arange(100)/10
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1)
plt.plot(x,y2,color='yellow',label="cosx")
plt.scatter(x,y1,color='blue',label="sinx")

plt.legend(loc="upper right")
plt.show()
```

```python
x=np.arange(100)/10
y=np.sin(x)
plt.plot(x,y,color='purple', marker="s", markersize=10,alpha=0.1)
```

```python
'''
plt.plot(x,y):
color:çizginin rengini belirler
linestyle: çizgi stilini belirler
marker:kesimşimleri belirler
markersize: yukarıdakinin size ı
alpha: saydamlık opaklık
label:etiketi belirler plt.legend(loc='right') konumunu belirttir
'''
```

```python
x=np.arange(100)/10
y=np.cos(x)
plt.scatter(x,y,color='red',marker='o',label="cosx",edgecolors='blue',linewidths=2,alpha=0.3)
plt.legend()
plt.show()
```
