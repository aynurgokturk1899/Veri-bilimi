MATPLOTLİB BARPLOT VE HİSTOGRAM MANTIĞI

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data={"yıl": np.arange(2010,2021),
     "satis": np.random.randint(100,500,11),
     "masraf": np.random.randint(50,400,11),
     }
df=pd.DataFrame(data)

regions=["kuzey","Güney","Doğu","Batı"]

df['bölge']= np.random.choice(["kuzey","Güney","Doğu","Batı"],df.shape[0])
df
```

```python
df
```

```python
plt.plot(df["yıl"],df["satis"],label="satışlar", linestyle="--",marker='o' , color="red",markerfacecolor="black")
plt.legend(loc="upper right")
plt.show()
```

```python
# yıl ve masrafı aynı plota ekleyin 
# marker :x
```

```python
plt.plot(df["yıl"],df["satis"],label="satışlar", linestyle="--",marker='o' , color="red",markerfacecolor="black")
plt.legend(loc="upper right")

plt.plot(df["yıl"],df["masraf"],label="yıllara göre masraf grafiği",marker="x", color="green",linestyle='--')
plt.legend(loc="upper right")
plt.title("yıllık satış ve masraflar",fontsize=16)
plt.xlabel("yıllar",fontsize=16)
plt.ylabel("TL",fontsize=16)
plt.show()
```

HİSTOGRAM

```python
df.groupby("bölge").sum() # bölge bazında satışı gösterir yili masrafı gösterir
```

```python
sales_by_region=df.groupby("bölge")["satis"].sum()
```

```python
sales_by_region
```

```python
sales_by_region.index
```

```python
plt.bar(sales_by_region.index,sales_by_region.values,label="bölgelerdeki toplam satışlar")
plt.xlabel("Bölge")
plt.ylabel("Satış")
plt.legend(loc="upper right")
```

```python
df
```

```python
sales_bins=pd.cut(df["satis"],bins=10) # kaçarlı gruplara ayırayım
```

```python
sales_bins.value_counts()  # iki fonksiyon var pd.cut ve pd.qcut
# pandas cut : bin arası mesafe
# bin: numerik datanın grubu olarak düşünülebilir
```

```python
sales_bins=pd.qcut(df["satis"],q=10)
```

```python
sales_bins
```

```python
bin_counts=sales_bins.value_counts().sort_index()
bin_counts
```

```python
bin_counts.index
```

```python
bin_middles=[(i.left+ i.right)/2 for i in bin_counts.index]
bin_middles
```

```python
plt.bar(bin_middles,bin_counts.values,width=300.3,alpha=0.3)
```

```python
x=np.random.randn(1000)
plt.hist(x,bins=30,color="black",range=(-10,10),density=True)
plt.show()
```

```python
# plt.hist numerik değerlerinizin dağılımınu yorumlayan birgrafiktir, histogram çizer
# bins: barlardan kaç tane olsun diyor
# range : binlerin alt ve üst limitini belirler
# density: True olursa grafik yoğunluk belirtir fa
```
