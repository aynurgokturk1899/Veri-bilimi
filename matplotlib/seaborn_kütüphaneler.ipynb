```python
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
```

```python
plt.rcParams["figure.figsize"]=(10,6)
plt.rcParams["xtick.labelsize"]=14
plt.rcParams["ytick.labelsize"]=14

plt.rcParams["axes.labelsize"]=16
```

```python
plt.rcParams["figure.figsize"]=(10,6)
```

```python
data= sns.load_dataset("penguins")
data
```

```python
# Seaborn olmadan histogram bulma
data_male=data[data['sex']=='Male']['flipper_length_mm']
data_Female=data[data['sex']=='Female']['flipper_length_mm']

sns.histplot(data=data_Female, label="Female",alpha=0.5)
sns.histplot(data=data_male, label="Male")
plt.legend(loc="right")
sns.histplot(data=data,x="flipper_length_mm")
```

```python
# seaborn ile histogram
sns.histplot(data=data,x='flipper_length_mm',hue='sex')
plt.show()
```

```python
grouped_data.index
```

```python
sns.barplot(
           x=grouped_data.index,  
           y=grouped_data.values)  # RESET.index demeden de ben bunları oluşturabiliriiim
```

```python
color_map=plt.get_cmap("viridis")
color_map
```

```python
norm=plt.Normalize(grouped_data.min(),grouped_data.max())
```

```python
norm([2])
```

```python
colors=color_map(norm(grouped_data.values))
colors
```

```python
fig, ax = plt.subplots(figsize=(12,9))

sns.barplot(
    x=grouped_data.index,
    y=grouped_data.values,
    palette=colors.tolist()
)

ax.set_ylabel("Average body_mass_g")

# Colorbar ekleme
cax = fig.add_axes([0.95, 0.25, 0.03, 0.5])  # 4 değer
sm = plt.cm.ScalarMappable(cmap=color_map, norm=norm)
sm.set_array([])  # gerekli olabilir
fig.colorbar(sm, cax=cax)

plt.show()
```
