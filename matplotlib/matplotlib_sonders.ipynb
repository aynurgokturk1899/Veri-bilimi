```python
df["type"].mode().iloc[0]       # bir sÃ¼tunun en Ã§ok tekrar eden deÄŸeri gÃ¶sterir
```

```python
for column in df.columns:
    print("column:",column)
    mask=df[column].isnull()

    if mask.any():
        print("\t null deÄŸerler var",column)
        mode_value=df[column].mode().iloc[0]   # bize seri dÃ¶ndrÃ¼r
        print("\t\t mode deÄŸerleri ile dolduruyor",mode_value)
        df[column].fillna(mode_value,inplace=True)
```
