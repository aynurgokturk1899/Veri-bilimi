```python
for column in df.columns:   # burada columnlara bakıyoruz null değerleri olan kaç satır varsa topluyor bir sayi söylüuor
    print("column:",column)
    print("toplam null değerler:", df[column].isnull().sum(),"\n")
```

```python
df["type"].mode().iloc[0]       # bir sütunun en çok tekrar eden değeri gösterir
```

```python
for column in df.columns:
    print("column:",column)
    mask=df[column].isnull()

    if mask.any():
        print("\t null değerler var",column)
        mode_value=df[column].mode().iloc[0]   # bize seri döndrür
        print("\t\t mode değerleri ile dolduruyor",mode_value)
        df[column].fillna(mode_value,inplace=True)
```

# Görselleştirmeler  

```python
sns.countplot(data=df, x="type")   # bunu barplot ile de yapabiliriz 
plt.title("içeriklerin türleri")
plt.show()
```

```python
df["type"].value_counts().plot(kind="pie",autopct="%1.1f")
plt.title("Film ve Dizilerin dağılımları")
plt.show()
```

```python
top_countries=df["country"].value_counts().head(11)
top_countries
```

```python
sns.barplot(x=top_countries.index, y=top_countries,palette="tab10")
plt.title("En çok 10 içerik üreten 10 ülke")
plt.xticks(rotation=60)
plt.show()
```
