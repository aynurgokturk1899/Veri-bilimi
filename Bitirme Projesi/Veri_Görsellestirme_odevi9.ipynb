```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Arial'
```

```python
data = pd.read_csv('50_Startups.csv')
```

# ## Bu veri seti, 50 farklı startup şirketinin çeşitli harcamalarını ve kârlılıklarını içermektedir.  
# 
# R&D Spend (Ar-Ge Harcaması): Şirketin araştırma ve geliştirme (Ar-Ge) için harcadığı tutar.  
# Administration (Yönetim Harcaması): Şirketin yönetim giderleri için harcadığı tutar.  
# Marketing Spend (Pazarlama Harcaması): Şirketin pazarlama ve reklam faaliyetleri için harcadığı tutar. 
# State (Eyalet): Şirketin faaliyet gösterdiği eyalet (örneğin, New York, California, Florida).  
# Profit (Kâr): Şirketin elde ettiği toplam kâr.  
# Bu veri seti, startup'ların çeşitli harcama kalemleri ile kârlılıkları arasındaki ilişkileri analiz   etmek için kullanılabilir. Örneğin, Ar-Ge veya pazarlama harcamalarının kârlılık üzerindeki etkisini   incelemek için uygun bir veri setidir.  

## ## 1.GÖREV : R&D Harcaması ve Kâr Arasındaki İlişki (Scatter Plot): Ar-Ge harcamaları ile kâr arasındaki ilişkiyi gösteren bir dağılım grafiği.


```python
plt.scatter(data['R&D Spend'],data['Profit'],label="Ar-Ge Harcamaları ve Kâr İlişkisi",color='red')
plt.legend(loc="upper left")

plt.xlabel("R&D Spend (Ar-Ge Harcamaları)")
plt.ylabel("Profit (Kâr)")
plt.title("Ar-Ge harcamaları ile kâr arasındaki ilişkiyi gösteren dağılım grafiği.")
plt.show()
```

## ## 2.GÖREV: Yönetim Harcamaları ve Kâr Arasındaki İlişki (Scatter Plot): Yönetim harcamaları ile kâr arasındaki ilişkiyi gösteren bir dağılım grafiği.

```python
plt.plot(data["Administration"],data["Profit"],label="Yönetim harcamaları ile kâr",
 linestyle="--",marker='o' , color="red",markerfacecolor="black")
plt.legend(loc="right")

plt.legend(loc="best")
plt.title("Yönetim harcamaları ile kâr arasındaki ilişkiyi gösteren bir dağılım grafiği.",fontsize=16)
plt.xlabel("Administration (Yönetim Harcaması)",fontsize=16)
plt.ylabel("Profit (Kâr)",fontsize=16)




plt.show()
```

# ## 3. GÖREV: Eyaletlere Göre Ortalama Kâr (Bar Chart): Farklı eyaletlerdeki startup'ların ortalama kârlarını karşılaştıranAbir çubuk grafik.


```python
state_profit_avg = data.groupby('State')['Profit'].mean()
plt.figure(figsize=(10,6))
state_profit_avg.plot(kind='bar', color='skyblue',label="Eyaletlere Göre Ortalama Kâr")
plt.legend(loc="upper right")


plt.xlabel("State (Eyalet)",fontsize=16)
plt.ylabel("Profit (Kâr)",fontsize=16)
plt.title("eyaletlerdeki startup'ların ortalama kârlarını karşılaştıranAbir çubuk grafik")

plt.xticks(rotation=45)
plt.show()
```
