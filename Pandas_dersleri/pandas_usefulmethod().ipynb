AMAÇ:

```python
# yeni sütun olacak fiyat kaegorisi 10 dolardan az ödediyse $ 10 ile 20  $$ 20 ile 25 $$$ 25 ile 30 arası $$$$ işareti bulunan kategori yap
```

```python
def fiyat_kategorisi(x):
    if x<10:
        return '$'
    elif 10<x<=20:
        return '$$'
    elif 20<x<30:
        return '$$$'
   
```

```python
df['total_bill'].apply(fiyat_kategorisi)
```

```python
df['fiyat_kategorisi']=df['total_bill'].apply(fiyat_kategorisi)
```

3.YOL VEKTORİZED

```python
df
```

```python
def fiyat_kategori_vektorized(bill):
    categories= np.full_like(bill, '$', dtype='U3' )

    categories[(bill>10) & (bill<20)] ='$$'

    categories[bill>=20] ='$$$' 

    return categories
    
```

```python
df['fiyat_kategorisi']= fiyat_kategori_vektorized(df['total_bill'].values)
```
